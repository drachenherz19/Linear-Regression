```python
import pandas as pd
import numpy as np
from google.colab import drive
import matplotlib.pyplot as plt

# Loading the data from Google Drive:
drive.mount('/content/drive')
file_path ='/content/drive/MyDrive/ML/HW2_linear_data-1.csv'
df = pd.read_csv(file_path)

# Extracting the X and Y values:
X = df.iloc[:, 0].values
Y = df.iloc[:, 1].values

# Initialize slope and y-intercept for Linear Regression:
m = 0
c = 0

# Defining Learning rate
L = 0.0001

# The number of iterations:
epochs = 1000

n = float(len(X))

# Performing Gradient Descent
for i in range(epochs):
    Y_pred = m*X + c
    D_m = (-2/n) * sum(X * (Y - Y_pred))
    D_c = (-2/n) * sum(Y - Y_pred)
    m = m - L * D_m
    c = c - L * D_c

# Prediction:
Y_pred = m*X + c

# Calculating Mean Squared Error:
mse = np.mean((Y - Y_pred)**2)

# Plot:
plt.scatter(X, Y, color='red', label='Data points')
plt.plot(X, Y_pred, color='blue', label='Linear regression line')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Return the parameters of the linear model:
(m, c, mse)
```
